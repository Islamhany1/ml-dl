import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

read the data

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
X
```

train the model

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)
```

create x to the power of something

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 4)
X_poly = poly_features.fit_transform(X)
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, Y)
X_poly.shape
```

visualise linear model

```python
plt.scatter(X, Y, color = 'r')
plt.plot(X, linear_regressor.predict(X), color = 'b')
plt.xlabel('level')
plt.ylabel('salary')
plt.title('truth or bluff (linear regression)')
plt.show()
```

visualise polynomial model

```python
plt.scatter(X, Y, color = 'r')
plt.plot(X, poly_regressor.predict(X_poly), color = 'b')
plt.xlabel('level')
plt.ylabel('salary')
plt.title('truth or bluff (poly linear regression)')
plt.show()
```

predict new value

```python
print(poly_regressor.predict(poly_features.fit_transform([[6.5]])))
linear_regressor.predict([[6.5]])
```
